```python
%matplotlib notebook
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
sns.scatterplot(df.petal_length,df.petal_width,hue=df.species)
```

```python
sns.pairplot(df,hue = 'species')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
df.columns
```

```python
data = scale(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
data = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
kmeans.fit(data)
```

```python
kmeans.labels_
```

```python
sns.scatterplot(df.petal_length,df.sepal_width,hue=df.species,style= kmeans.labels_)
```
